```python
#imports all csv files in folder and concatonates the data sets from trackmate.

import glob

df = pd.DataFrame()
for filename in glob.glob(data+'*.csv'):
    data_01 = pd.read_csv(filename, sep=',')
    filename = filename.split('/')[6]
    print(filename)
    filename = filename.split('.')[0]
    print(filename)
    #well = filename.split('_')[6]
    data_01['Construct'] = 'AScIELVA'
    data_01['Time'] = '24hpi'
    print(filename.split('_')[2])
    data_01['Select_ch'] = filename.split('_')[2]
    #data_01['Inclusion'] = int(filename.split('_')[2])
    data = data_01[['Select_ch','Construct','Time','Track_ID','Frame', 'X', 'Y', 'Channel_1', 'Channel_2', 'Channel_3']]
    df = df.append(data,  ignore_index=True)
df['Sample_ID'] = df.Construct + '_'+ df.Select_ch
#df['Sample_ID'] = df.Sample_ID + '_'+ df.Time
#df['Sample_ID'] = df.Sample_ID + '-Track-' + df.TRACK_ID.astype(str)
```

```python
# Create dataframe of maximum expression of individual cells

full_max_df = None
for track in df['Track_ID'].unique():
    
    this_df = df[df['Track_ID'] == track]
    max_chan1 = this_df['Channel_1'].max()
    
    this_df = this_df[this_df['Channel_1'] == max_chan1]
    
    if full_max_df is None:
        full_max_df = this_df.copy(deep=True)
    else:
        full_max_df = pd.concat([full_max_df, this_df])
```
